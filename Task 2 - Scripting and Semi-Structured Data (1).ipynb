The text file comes in JSON format and here are some keys and their description. They are only the most important ones for this task.

|key| description |
|---|-----------|
| a |Denotes information about the web browser and operating system|
| tz | time zone |
| r | URL the user come from |
| u | URL where the user headed to |
| t | Timestamp when the user start using the website in UNIX format |
| hc | Timestamp when user exit the website in UNIX format |
| cy | City from which the request intiated |
| ll | Longitude and Latitude |

```python
# importing dependancies 
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
json = pd.read_json('usa.gov_click_dat.json', lines=True)
```

```python
json
```

```python
json.dtypes
```

```python
# cgange column C name to City
json.rename(columns={'cy':'City'}, inplace=True)
json.head(3)
```

```python
# get the Browser name from string by spliting text by '/' 
json[['Browser','Operating_system','well_delet']] = json['a'].str.split(n=2,expand=True)
json.head(3)
```

```python
# remove number from string using regular exp '\d'
json['Browser']=json['Browser'].str.replace('\d+','', regex=True)
json.head(3)
```

```python
json[['remov1','remove3','from_url', 'remove4']]=json['r'].str.split('/', n=3, expand =True)
json.head(3)
```

```python
json[['remov6','remove4','to_url', 'remove5']]=json['u'].str.split('/', n=3, expand =True)
json.head(3)
```

```python
json[['Continent','Time_zone']]=json['tz'].str.split('/', n=2,expand=True)
```

```python
json['time_in']=pd.to_datetime(json['t'],unit='s').dt.time
```

```python
# this column has unix patren for date , and there is 1 value have over 1 ,so we need to remove it 
json['hc'].replace(13340070364,3340070364,inplace=True)
```

```python
# convert from unix format to date format and extract time 
json['time_out']=pd.to_datetime(json['hc'],unit='s').dt.time
```

```python
json.head()
```

```python
# remove unwanted columns
df=json[['Browser','Operating_system','from_url','to_url','Time_zone','Continent','City','time_in','time_out','ll']]
```

```python
df
```

```python
# starting column name by upper case
df.columns=df.columns.str.capitalize()
```

```python
# remove '\.' from browser column
df['Browser']=df['Browser'].str.replace('/.','',regex=True)
```

```python
# remove '(' 
df['Operating_system']=df['Operating_system'].str.replace('(','',regex=True)
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# extract langtude from ll
df['Langitude']=df['Ll'].str.get(0)
```

```python
# extract latitude 
df['Latitude']=df['Ll'].str.get(1)
```

```python
df.drop('Ll',inplace=True, axis =1)
```
